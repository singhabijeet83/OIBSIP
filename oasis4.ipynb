```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string
string.punctuation
```

```python
#Lower Casing
def lower(text1):
  text1=text1.lower()
  return text1
lower('This Is an ExamPLe')
```

```python
#Tokenization
def token(text2):
  return nltk.word_tokenize(text2)
token('This is an examPLe')
```

```python
#Remove special Character
def remove_sp(text3):
  y=[]
  for i in text3:
    if i.isalnum():
      y.append(i)
  return y
remove_sp('Thi5 is @n examP1e')
```

```python
#Stopword and Punctuattion Reomoval
def remove_punc(text4):
  text4=lower(text4)
  text4=token(text4)
  y=[]
  for i in text4:
    if i.isalnum():
      y.append(i)
  text4=y[:]
  y.clear()

  for i in text4:
    if i not in string.punctuation and i not in stopwords.words('english'):
      y.append(i)
  return y
remove_punc('This is an examPLe 34 @#$ bye')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
ps.stem('loving')
```

```python
#Stemming
def stemming(text5):
  text5=lower(text5)
  text5=token(text5)
  y=[]
  for i in text5:
    if i.isalnum():
      y.append(i)
  text5=y[:]
  y.clear()
  for i in text5:
    if i not in string.punctuation and i not in stopwords.words('english'):
      y.append(ps.stem(i))
  text5=y[:]
  y.clear()
  for i in text5:
    y.append(ps.stem(i))
  return y
stemming('I am learning nEw Th1ngs While C0din6')
```

```python
def transform_text(text):
  text=lower(text)
  text=token(text)
  y=[]
  for i in text:
    if i.isalnum():
      y.append(i)
  text=y[:]
  y.clear()
  for i in text:
    if i not in string.punctuation and i not in stopwords.words('english'):
      y.append(ps.stem(i))
  text=y[:]
  y.clear()
  for i in text:
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['transformed_text']=df['v2'].apply(transform_text)
df.head()
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```
